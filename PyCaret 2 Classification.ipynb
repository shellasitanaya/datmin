# 1. Data Repository

```python
#Load dataset yang ada
from pycaret.datasets import get_data
index = get_data('index')
```

```python
#mengambil dataset juice
data = get_data('juice')
```

# 2. Initialize Setup

```python
#setup the classification environment
#membagi dataset menjadi data latih dan data uji
from pycaret.classification import *
clf1 = setup(data, target = 'Purchase', session_id=123, log_experiment=True, experiment_name='juice1')
```

# 3. Compare Baseline

```python
#membandingkan semua model yang ada di pycaret
best_model = compare_models()
```

# 4. Create Model

```python
lr = create_model('lr')
```

```python
dt = create_model('dt')
```

```python
rf = create_model('rf', fold = 5)
```

```python
models()
```

```python
models(type='ensemble').index.tolist()
```

```python
ensembled_models = compare_models(whitelist = models(type='ensemble').index.tolist(), fold = 3)
```

# 5. Tune Hyperparameters

```python
#mencari kombinasi hyperparameter terbaik
tuned_lr = tune_model(lr)
```

```python
tuned_rf = tune_model(rf)
```

# 6. Ensemble Model

```python
bagged_dt = ensemble_model(dt)
```

```python
#boosting mengumpulkan pembelajar lemah menjadi pembelajar kuat
boosted_dt = ensemble_model(dt, method = 'Boosting')
```

# 7. Blend Models

```python
#mengambil model yang sudah diboosting dan menggabungkan prediksi probabilitasnya untuk keputusan akhir
blender = blend_models(estimator_list = [boosted_dt, bagged_dt, tuned_rf], method = 'soft')
```

# 8. Stack Models

```python
#stacking untuk belajar dari prediksi model" sebelumnya
stacker = stack_models(estimator_list = [boosted_dt,bagged_dt,tuned_rf], meta_model=rf)
```

# 11. AutoML()

```python
best = automl(optimize = 'Recall')
best
```

# 12. Predict Model

```python
#Label = 0 'CH'
#Label = 1 'MM'
pred_holdouts = predict_model(lr)
pred_holdouts.head()
```

```python
new_data = data.copy()
new_data.drop(['Purchase'], axis=1, inplace=True)
predict_new = predict_model(best, data=new_data)
predict_new.head()
```
